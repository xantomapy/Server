```python
import cv2
import numpy as np
import random as r
from google.colab import drive

def draw_geometric_shapes(img):
    rectangle_pt1 = img
    i = 0
    while i <= 2:
      rectangle_pt1 = (r.randrange(400), r.randrange(400))
      rectangle_pt2 = (r.randrange(400), r.randrange(400))
      rectangle_color = (r.randrange(255),r.randrange(255),r.randrange(255))
      processed_img = cv2.rectangle(img, rectangle_pt1, rectangle_pt2, rectangle_color, -1)

      triangle_pt1 = (r.randrange(400), r.randrange(400))
      triangle_pt2 = (r.randrange(400), r.randrange(400))
      triangle_pt3 = (r.randrange(400), r.randrange(400))
      triangle_color = (r.randrange(255),r.randrange(255),r.randrange(255))
      triangle_cnt = np.array( [triangle_pt1, triangle_pt2, triangle_pt3] )
      cv2.drawContours(processed_img, [triangle_cnt], 0, triangle_color, -1)

      line_pt1 = (r.randrange(400), r.randrange(400))
      line_pt2 = (r.randrange(400), r.randrange(400))
      line_color = (r.randrange(255),r.randrange(255),r.randrange(255))
      processed_img = cv2.line(processed_img, line_pt1, line_pt2, line_color, 9)
      i += 1;

    print('Geometric shapes was drawed on the image')
    return processed_img

drive.mount("/content/drive")
original_img = cv2.imread('./drive/My Drive/VPython/img_python_labs/lab_1/my_img.png')

img_task_1 = draw_geometric_shapes(original_img)
cv2.imwrite('./drive/My Drive/VPython/img_python_labs/lab_1/img_task_1.png', img_task_1)
```

```python
def paint_picture_gray(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

img_task_2 = paint_picture_gray(img_task_1)
cv2.imwrite('./drive/My Drive/VPython/img_python_labs/lab_1/img_task_2.png', img_task_2)
```

```python
def mirror_picture(img):
    return cv2.flip(img, 1)

img_task_3 = mirror_picture(img_task_1)
cv2.imwrite('./drive/My Drive/VPython/img_python_labs/lab_1/img_task_3.png', img_task_3)
```

```python
def set_rotate_90_degress(img):
    return cv2.rotate(img, cv2.cv2.ROTATE_90_CLOCKWISE)

img_task_4 = set_rotate_90_degress(original_img)
cv2.imwrite('./drive/My Drive/VPython/img_python_labs/lab_1/img_task_4.png', img_task_4)
```

```python
def decrease_picture(img):
    scale_percent = 50
    width = int(img.shape[1] * scale_percent / 100)
    height = int(img.shape[0] * scale_percent / 100)

    dsize = (width, height)
    return cv2.resize(img, dsize)

img_task_5 = decrease_picture(original_img)
cv2.imwrite('./drive/My Drive/VPython/img_python_labs/lab_1/img_task_5.png', img_task_5)
```

```python
def set_rotate_30_degress(img):
    img_center = tuple(np.array(img.shape[1::-1]) / 2)
    rot_mat = cv2.getRotationMatrix2D(img_center, 30, 1.0)
    return cv2.warpAffine(img, rot_mat, img.shape[1::-1], flags=cv2.INTER_LINEAR)

img_task_6 = set_rotate_30_degress(original_img)
cv2.imwrite('./drive/My Drive/VPython/img_python_labs/lab_1/img_task_6.png', img_task_6)
```
