```python
import cv2
import numpy as np
from google.colab import drive
drive.mount("/content/drive")
img = cv2.imread('./drive/My Drive/VPython/img_python_labs/lab_1/img_task_5.png')
img[img==255]=254
img2=img


n=img2.shape[0]*img2.shape[1]*img2.shape[2]
for i in range(0,img2.shape[0]):
  for j in range(0,img2.shape[1]):
    color=img2[i,j]
    r=0
    for jj in range(0,color[0]):
      nr=np.count_nonzero(img2 == jj)
      r=r+nr/n
    r*=255
    
    g=0
    for jj in range(0,color[1]):
      ng=np.count_nonzero(img2 == jj)
      g=g+ng/n
    g*=255

    b=0
    for jj in range(0,color[2]):
      nb=np.count_nonzero(img2 == jj)
      b=b+nb/n
    b*=255
    img2[i,j]=(r,g,b)
cv2.imwrite('./drive/My Drive/VPython/img_python_labs/lab_3/img_task_1.png',img2)
```

```python

import cv2
import numpy as np
from google.colab import drive
drive.mount("/content/drive")
img = cv2.imread('./drive/My Drive/VPython/img_python_labs/lab_3/img_task_1.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
equ = cv2.equalizeHist(img_gray)
cv2.imwrite('./drive/My Drive/VPython/img_python_labs/lab_3/img_task_2.png',equ)

```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
drive.mount("/content/drive")
img = cv2.imread('./drive/My Drive/VPython/img_python_labs/lab_3/img_task_1.png')
color = ('b','g','r')
for i,col in enumerate(color):
  histr = cv2.calcHist([img],[i],None,[256],[0,256])
  plt.plot(histr,color = col)
  plt.xlim([0,256])
plt.show()
```
